Day 38 Python Challenge  String Methods and Slicing

Project Title: Feedback Analyzer

Scenario:

A content review team wants to analyse user feedback and comments. They want to check how frequently certain words appear and extract parts of the feedback for reporting.

Tasks:



```python




# 1. Ask the user to input a sentence
feedback = input("Enter your feedback: ")


# 2. Count how many words are in the sentence
word_count = len(feedback.split())
print("Total number of words:", word_count)


# 3. Count how many times the word “data” appears (case-insensitive)
data_count = feedback.lower().count("data")
print("Number of times 'data' appears:", data_count)


# 4. Extract first 5 and last 5 characters
first_five = feedback[:5]
last_five = feedback[-5:]
print("First 5 characters:", first_five)
print("Last 5 characters:", last_five)


# 5. Reverse the sentence
reversed_feedback = feedback[::-1]
print("Reversed sentence:", reversed_feedback)

```
